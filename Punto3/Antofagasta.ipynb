```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
import csv
from scipy import stats
import scipy.linalg as la
import math
%pylab inline
```

```python
#Importar Datos (no se importan los datos de errores ya que se omitirá este problema al ser una primera aproximación)
datos = np.loadtxt("antofagasta.dat")
latitud   = datos[:,0]
longitud   = datos[:,1]
uEast    = datos[:,2]
uNorth   = datos[:,4]

#Grados a Radianes
latitud =  latitud*(np.pi/180)
longitud = longitud*(np.pi/180)

#Tamaño de los vectores
length  = latitud.shape[0]

#inicializar varibles importantes
p1 = 0
p2 = 0
p3 = 0
p4 = 0
p5 = 0

#Coordenadas de Lat-Lon a UTM, usando relaciones entre los sistemas geodéticos y constantes propuestas
#Aquí se establece el sistema de referencia común
po = 0.9996
c  = 0.007
e  = 0.08

a  = 6378137
b  = 6356752.3142
n  = (a-b)/(a+b)


#Funciones que determinan las conversiones ncesarias para pasar a UTM
def conversion2(k):
    kk = a*((1-e**2/4-3*e**4/64-5*e**6/256)*k-(3*e**2/8+3*e**4/32+45*e**6/1024)*np.sin(2*k)+(15*e**4/256 + 45*e**6/1024)*np.sin(4*k) - (35*e**6/3072)*np.sin(6*k))
    return kk

def conversion(k):
    kk = a/((1-e**2*(np.sin(k))**2)**(1/2))
    return kk


#Inicializar verctores
latitudUTM = np.zeros(length)
longitudUTM = np.zeros(length)

#Conversion final a UTM
for i in range (0, length):
    lat = latitud[i]
    p2=po*conversion(lat)*np.sin(lat)*np.cos(lat)*0.5
    p3=(po*conversion(lat)*np.sin(lat)*(np.cos(lat))**3*(1/24))*(5-(np.tan(lat))**2+9*c*(np.cos(lat))**2+4*c**2*(np.cos(lat))**4)
    p5=(po*conversion(lat)*(np.cos(lat))**3*(1/6))*(1-(np.tan(lat)**2)+c*(np.cos(lat))**2)
        
    ppp = longitud[i]
    l = 291*(np.pi/180)
    p = ppp -l
    
    v1=po*(conversion2(lat))+(p2*p**2)+(p3*p**4)
    v2=(po*conversion(lat)*np.cos(lat)*p)+(p5*p**3)
    
    latitudUTM[i] = v1
    longitudUTM[i] = v2
                                                            
```

```python
#The following process was used using the proposed procedure in the book of Allmendinger, R. W., 2012. Strcutural Geology Algorithms. 
#Cambridge. 144-145.

#Mean vector, overall orientations
UE = 0
UN = 0
for i in range (0, length):
    UE = UE + uEast[i]
    UN = UN + uNorth[i]
    
UE=UE/length
UN=UN/length

#Transformation matrix for X1'
t11 = np.arctan(UE/UN)
t12 = (np.pi)/2-t11
t22 = t11
t21 = (np.pi)/2+t22
a11 = np.cos(t11)
a12 = np.cos(t12)
a21 = np.cos(t21)
a22 = np.cos(t22)

#Transform the east and north coordinates of the GPS vectors
latTrans = np.zeros(length)
lonTrans = np.zeros(length)
for i in range (0, length):
    v1=latitudUTM[i]
    v2=longitudUTM[i]    
    v1p = v1*a11+v2*a12
    v2p = v2*a21+v2*a22
    latTrans[i] = v1p
    lonTrans[i] = v2p
```

```python
# determine best fit line for displacement in East
slope, intercept, r_value, p_value, std_err = stats.linregress(pos, magvalor)

print 'r value', r_value
print  'p_value', p_value
print 'standard deviation', std_err

line = slope*pos+intercept
plot(pos,line,'r-',pos, magvalor,'o')
```

Strain varies in two dimensions because gradients in displacement exist in two dimensions. There is clearly some stations with higher displacement, which are the ones located aproximately in the same area and closer to the epicenter. Once the positions of the stations get farther away, the magnitude of displacement tends to zero as can be spected as stress and strain have been accomodated in the lithosphere. If the scatter and the regression is done just over the highest displacement stations, the linear regression coeficient will get closer to 1, indicating that the strain can be effectively correlated to the earthquake results.
Worth to take notice is the low magnitude values of 2 GPS stations with same distance as the biggest magnitude displacement stations; since the distance is taken from a common center point (UTM), those stations lie over a great circle and so, don't have to be over the earthquake epicenter but to the north or south of it.
