```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load image
img = "vesicula_dic001-1.tif"
Im = plt.imread(img)
fig = plt.figure(1, figsize=(10.5,8.5))
plt.imshow(Im[:,:],cmap='gray')
plt.title('Original Image', fontsize=12)
[x_axis,y_axis] = shape(Im)
```

```python
maxA = amax(Im)
minA = amin(Im)
Im1 = zeros((x_axis,y_axis))
Im1 = ((1.0*Im-minA)/maxA)

plt.imshow(Im1[:,:],cmap='gray')
plt.title('Normalized Image', fontsize=12)
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return -chi_squared
```

```python
def my_model(Im, r_0, r_1, m, phi):
    theta = linspace(0,2*pi,10000)
    r = r_0 + r_1 * sin(m * theta + phi)
    x = r * cos(theta)
    y = r * sin(theta)
    [x_axis,y_axis] = shape(Im)
    Im1 = zeros(shape(Im))
    Im1[int_(x), int_(y)] = 1.0        
    return Im1
```

```python
r0_walk = empty((0)) #this is an empty list to keep all the steps
r1_walk = empty((0))
m_walk = empty((0))
phi_walk = empty((0))
l_walk = empty((0))

r0_walk = append(r0_walk, (x_axis+y_axis)*random.random()/4.0)
r1_walk = append(r1_walk, random.random())
m_walk = append(m_walk, random.randint(0,5))
phi_walk = append(phi_walk, 2*pi*random.random())

Image_init = my_model(Im, r0_walk[0], r1_walk[0], m_walk[0], phi_walk[0])
l_walk = append(l_walk, likelihood(Im1, Image_init))
print r0_walk
print r1_walk
print m_walk
print phi_walk
print l_walk
plt.imshow(my_model(Im,r0_walk[0],r1_walk[0],m_walk[0],phi_walk[0]),cmap='gray')
plt.title('Primera iteracion', fontsize=12)
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    r0_prime = np.random.normal(r0_walk[i], 0.5) 
    r1_prime = np.random.normal(r1_walk[i], 0.1)
    m_prime = np.random.normal(m_walk[i], 0.1)
    phi_prime = np.random.normal(phi_walk[i], 0.1)

    Image_init = my_model(Im, r0_walk[i], r1_walk[i], m_walk[i], phi_walk[i])
    Image_prime = my_model(Im, r0_prime, r1_prime, m_prime, phi_prime)
    
    l_prime = likelihood(Im1, Image_prime)
    l_init = likelihood(Im1, Image_init)
    
    alpha = exp(l_prime - l_init)
    if(alpha>=1.0):
        r0_walk  = append(r0_walk, r0_prime)
        r1_walk  = append(r1_walk, r1_prime)
        m_walk  = append(m_walk, m_prime)
        phi_walk  = append(phi_walk, phi_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            r0_walk  = append(r0_walk, r0_prime)
            r1_walk  = append(r1_walk, r1_prime)
            m_walk  = append(m_walk, m_prime)
            phi_walk  = append(phi_walk, phi_prime)
            l_walk = append(l_walk, l_prime)
        else:
            r0_walk = append(r0_walk, r0_walk[i])
            r1_walk = append(r1_walk, r1_walk[i])
            m_walk = append(m_walk, m_walk[i])
            phi_walk = append(phi_walk, phi_walk[i])
            l_walk = append(l_walk, l_init)
```
